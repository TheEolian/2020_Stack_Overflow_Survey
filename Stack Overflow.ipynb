### Filtering - Using Conditionals to Filter Rows and Columns

```python
# Create a filter variable
high_salary = (df['ConvertedComp'] > 70000)
```

```python
# Return DF with all rows that meet the high_salary criteria
df.loc[high_salary]
```

```python
# Return DF with all rows that DO NOT meet the high_salary criteria
df.loc[~high_salary]
```

 ### What percentage of people each each country work with python?

```python
# Create variable for number of respondents from each country
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
# Create variable for number of respondents who know Python from each country
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
# Concat the two series together
# Set axis equal to columns to it colcatinates on the indexes not on rows
python_df = pd.concat([country_respondents, country_uses_python], axis ='columns', sort=False)
python_df
```

```python
# Rename columns in context of what we are trying to do
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'KnowsPython'}, inplace=True)
python_df
```

```python
# Create a new column for the percent of people in the country who know Python
python_df['PctKnowsPython'] = (python_df['KnowsPython']/python_df['NumRespondents']) * 100
```

```python
python_df
```

```python
# Sort by percent of contry respondents who know python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```

## Cleaning Data - Casting Datatypes and Handling Missing Values
